```python
def build_SA(S, A):
    _, m = S.shape
    x = np.arange(m)
    # x will be broadcast to the size of S
    return A[S, x]
```

```python
from PIL import Image
img = Image.open('image.png').convert('LA')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])

img = mpimg.imread('image.png')     
gray = rgb2gray(img)    
plt.imshow(gray, cmap = plt.get_cmap('gray'))
plt.show()
```

```python
S, A = lc.local_clustering_l2(gray, 3, 10, nbiter=100)
```

```python
plt.imshow(A, cmap = plt.get_cmap('gray'), interpolation='nearest', aspect='auto')
```

```python
plt.imshow(build_SA(S, A), cmap = plt.get_cmap('gray'), interpolation='nearest', aspect='auto')
```

```python
plt.imshow(S, interpolation='nearest', aspect='auto')
```

```python
plt.imshow(build_SA(S, A), cmap = plt.get_cmap('gray'), interpolation='nearest', aspect='auto')
```
